```python
# user menu
print("-----------------------------------------")
print("-Welcome!                               -")
print("-Please Choose What You Want To Do.     -")
print("-Add Event Enter 1                      -")
print("-Remove Event Enter 2                   -")
print("-View Events Enter 3                    -")
print("-Search For Event Enter 4               -")
print("-Exit Enter 5                           -")
print("-----------------------------------------")


# Project Requirement data type 1
numberOfEvents=0
# Project Requirement data type 2 it receives it as a text
serviceNum = input()

# Project Requirement list
#event list
eventList = []

# Project Requirement lambda function 
# lambda function to print the eventList
viewEvents = lambda : print(eventList)


# Project Requirement function return
# function search for an event
def searchEvent(searchEventt): 
# Project Requirement data type 3    
    found = False
#     print("-What is the Event you are looking for!")
#     searchEventt = input()
    for i in eventList:
        if searchEventt == i:                
            found = True
    return found


# Project Requirement loop    
# it loops until the user exits 
while(serviceNum != 5):
# Project Requirement conditions
#     add event
    if serviceNum == '1': 
        print("-Please give me the name of the event to add! ")
        addEvent = input()
#       eventList.extend(addEvent) it stores the text by each character
#       eventList.insert(addEvent) it insert a value in specified index
        eventList.append(addEvent)
        numberOfEvents += 1
        print("-The number of events you have is ",numberOfEvents)
        print("-Anything else you want to do?")
        serviceNum = input()
        
# remove event        
    elif serviceNum == '2':
        print("-Please give me the name of the event to remove! ")
        removeEvent = input()
        flag = searchEvent(removeEvent)
        if flag:
            eventList.remove(removeEvent)
            numberOfEvents -= 1
            print("-Event removed!")
        else:
            print("-No event with the given name!")
        print("-The number of events you have is ",numberOfEvents)
        print("-Anything else you want to do?")
        serviceNum = input()

# view events
    elif serviceNum == '3':
        viewEvents()
        print("-Anything else you want to do?")
        serviceNum = input()

# search for event
    elif serviceNum == '4':
        print("-What is the Event you are looking for!")
        searchEventt = input()
        flag = searchEvent(searchEventt)
        if flag:
            print("-Yes you have this event")
        else:
            print("-No event with such name")
        print("-Anything else you want to do?")
        serviceNum = input()

# exit the loop
    elif serviceNum == '5':
        break
    

# user enters wrong number 
    else:
        print("-Please enter proper service number!")
        serviceNum = input()
```
